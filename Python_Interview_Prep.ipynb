## Блок 1: Базовые типы и Изменяемость (Mutability)

### 🔎 Теория
В Python все является объектом. Важное деление объектов — на **изменяемые (mutable)** и **неизменяемые (immutable)**.

#### 1. Immutable (Неизменяемые)
- **`int`, `float`, `complex`**: Числа. Операции: арифметика (`+`, `-`, `*`, `/`, `//`, `%`, `**`), побитовые для int (`&`, `|`, `^`, `~`, `<<`, `>>`).
- **`str`**: Строки (Unicode). Операции: конкатенация (`+`), дублирование (`*`), срезы, методы (`split`, `join`, `find`, `replace`, `strip`).
- **`tuple`**: Кортежи. Операции: доступ по индексу, срезы, `count()`, `index()`. Часто используются для фиксации набора данных.
- **`frozenset`**: Неизменяемое множество.
- **`bytes`**: Неизменяемая последовательность байтов.

#### 2. Mutable (Изменяемые)
- **`list`**: Списки. Операции: `append`, `extend`, `insert`, `remove`, `pop`, `sort`, `reverse`, срезы.
- **`dict`**: Словари (хэш-таблицы). Операции: `get`, `keys`, `values`, `items`, `update`, `pop`, `popitem`.
- **`set`**: Множества. Операции: `add`, `remove`, `discard`, `pop`, `clear`, `union` (`|`), `intersection` (`&`), `difference` (`-`).
- **`bytearray`**: Изменяемый аналог `bytes`.

**Важно:** При изменении `mutable` объекта его `id` (адрес в памяти) остается прежним. При попытке изменить `immutable` создается новый объект с новым `id`.

#### ⚖️ Сравнение типов:
- **Словарь (`dict`) vs Множество (`set`):**
  - Оба используют хэш-таблицы (средняя скорость доступа O(1)).
  - `dict` хранит пары `ключ: значение`, `set` — только уникальные ключи.
  - В `dict` ключами могут быть только хэшируемые (immutable) объекты.
- **Список (`list`) vs Кортеж (`tuple`):**
  - `list` — изменяемый (mutable), `tuple` — нет (immutable).
  - `tuple` занимает меньше памяти и работает быстрее при итерации.
  - `tuple` может быть ключом словаря (если его элементы тоже неизменяемы).

````python
# Пример изменяемости
a = [1, 2, 3]
b = a
b.append(4)
print(f"Список a: {a} (id: {id(a)})") # a изменился тоже!
````

## Блок 2: Срезы, Comprehensions и Генераторы

- **Срезы:** `lst[start:stop:step]`. Позволяют эффективно получать части последовательностей. Помните: `start` включительно, `stop` — нет.
- **Генераторы:** объекты, которые поддерживают протокол итерации. Они не хранят все значения в памяти, а вычисляют следующее только при запросе.

### ⚡ Основные функции и методы генераторов:
1. **`yield`**: приостанавливает выполнение функции и возвращает значение. При следующем вызове выполнение продолжается с того же места.
2. **`next(gen)`**: запрашивает следующее значение у генератора. Если значений больше нет, выбрасывает `StopIteration`.
3. **`.send(value)`**: возобновляет работу генератора и передает значение в выражение `yield`.
4. **`.throw(type)`**: выбрасывает исключение внутри генератора.
5. **`.close()`**: завершает работу генератора.

````python
# List comprehension vs Generator
n = 1000
list_comp = [x**2 for x in range(n)]
gen_exp = (x**2 for x in range(n))
print(type(gen_exp))
````

## Блок 3: Функции: Аргументы, Замыкания и Области видимости

### 1. *args и **kwargs
- `*args` (non-keyword arguments): пакует произвольное число позиционных аргументов в кортеж.
- `**kwargs` (keyword arguments): пакует произвольное число именованных аргументов в словарь.

### 2. Замыкание (Closure)
Функция, определенная внутри другой функции, которая запоминает переменные из внешней области видимости даже после того, как внешняя функция завершила выполнение.

### 3. Область видимости (Scope) — LEGB Rule
Python ищет переменные в следующем порядке:
1. **L (Local)**: Внутри функции.
2. **E (Enclosing)**: В локальной области видимости объемлющих функций (от ближайшей к самой дальней).
3. **G (Global)**: На уровне модуля.
4. **B (Built-in)**: Специфические для Python имена (len, print и т.д.).

### 4. Аннотации типов (Type Annotations)
Позволяют указывать ожидаемые типы аргументов и возвращаемого значения. Полезно для читаемости и статического анализа (mypy).
```python
def add(a: int, b: int) -> int: return a + b
```

````python
def power(exponent):
    def inner(base):
        return base ** exponent
    return inner

square = power(2)
print(square(5)) # 25
````

## Блок 5: ООП (Инкапсуляция, MRO, super)

### 1. Инкапсуляция и уровни доступа
- **Public** (`name`): доступен отовсюду.
- **Protected** (`_name`): сигнализирует, что атрибут предназначен для внутреннего использования в классе и его подклассах. Это лишь соглашение (convention), технически доступ остается открытым.
- **Private** (`__name`): использует механизм **Name Mangling**. Имя преобразуется в `_ClassName__name`, что затрудняет случайный доступ или переопределение в дочерних классах. 

### 2. MRO (Method Resolution Order)
Алгоритм поиска метода в цепочке наследования. В Python используется **C3 Linearization**. Посмотреть порядок можно через `MyClass.mro()`.

### 3. super()
Динамически ищет следующий класс в цепочке MRO. Важно использовать в конструкторах при множественном наследовании.

### 4. Фабричные методы (Factory Methods)
Это альтернативные конструкторы, обычно помечаемые декоратором `@classmethod`. 
- Позволяют создавать объекты на основе разных входных данных (например, `from_json`, `from_csv`).
- Скрывают логику сложной инициализации внутри себя.

````python
class Base:
    def say(self): print("Base")

class childA(Base):
    def say(self): print("A"); super().say()

class childB(Base):
    def say(self): print("B"); super().say()

class GrandChild(childA, childB):
    def say(self): print("Grand"); super().say()

GrandChild().say() # Обратите внимание на порядок!
````

## Блок: Декораторы @staticmethod, @classmethod, @property

### 🔎 Теория
1. **`@staticmethod`**: обычная функция внутри класса. Не получает ни `self`, ни `cls`. Используется для логики, связанной с классом, но не зависящей от состояния.
2. **`@classmethod`**: получает первым аргументом класс (`cls`). Часто используется как фабричный метод.
3. **`@property`**: позволяет обращаться к методу как к атрибуту (getter). Можно также определить setter и deleter.

```python
class A:
    @property
    def name(self): return "A"
    
    @classmethod
    def create(cls): return cls()
    
    @staticmethod
    def info(): print("Static")
```

## Блок 6: Спец-методы (Dunder methods)

`__str__` vs `__repr__`:
- `__str__` — для пользователя (читаем).
- `__repr__` — для разработчика (отладка, должен быть однозначным).

````python
class Person:
    def __init__(self, name): self.name = name
    def __repr__(self): return f"Person('{self.name}')"
    def __str__(self): return f"Человек по имени {self.name}"

p = Person("Иван")
print(str(p))
print(repr(p))
````

## Блок 7: Алгоритмы (Бинарный поиск и Сортировка)

- **Бинарный поиск:** O(log n). Массив должен быть отсортирован.
- **Сортировка пузырьком:** O(n^2). Простейшая, но неэффективная.

````python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target: return mid
        elif arr[mid] < target: low = mid + 1
        else: high = mid - 1
    return -1

data = [1, 3, 5, 7, 9, 11]
print(f"Индекс 7: {binary_search(data, 7)}")
````

## Блок 8: Структуры данных (Стек, Очередь, Хеш-таблица)

#### 1. Стек (Stack)
- **Принцип**: LIFO (Last In, First Out).
- **Реализация**: `list` (`append`, `pop`).
- **Операции**: `push` (O(1)), `pop` (O(1)), `peek` (O(1)).

#### 2. Очередь (Queue)
- **Принцип**: FIFO (First In, First Out).
- **Реализация**: `collections.deque` (`append`, `popleft`).
- **Операции**: `enqueue` (O(1)), `dequeue` (O(1)).

#### 3. Хеш-таблица (Hash Table)
- **Суть**: Сопоставление ключа и значения через хеш-функцию.
- **Реализация**: `dict` и `set` в Python.
- **Сложность**: Поиск, вставка и удаление в среднем **O(1)**.
- **Коллизии**: Ситуация, когда разным ключам соответствует один хеш.

#### 4. Метод открытой адресации (Open Addressing)
Один из способов борьбы с коллизиями (используется в Python dict).
- **Суть**: Если ячейка занята, алгоритм ищет следующую свободную ячейку в самой таблице (линейное пробирование, квадратичное или двойное хеширование).
- **Преимущество**: Экономия памяти (не нужны списки/цепочки), но таблица может переполниться.

````python
from collections import deque

queue = deque(["A", "B", "C"])
queue.append("D")
print(queue.popleft()) # "A"
print(queue) # ["B", "C", "D"]
````

````python
#1. 
def is_balanced(s):
    """
    Классическая реализация проверки баланса скобок.
    """
    stack = []
    pairs = {'(': ')', '[': ']', '{': '}'}
    
    for char in s:
        if char in pairs:  # открывающая скобка
            stack.append(char)
        elif char in pairs.values():  # закрывающая скобка
            if not stack or pairs[stack.pop()] != char:
                return False
    
    return not stack
test = "{([])}"   
result  = is_balanced(test)
status = "✅" if result else "❌"
print(f"{status} {test:5} -> {result}")
````

````python
#2.
import random

"""
ПРОБЛЕМА ВЫБОРА ОПОРНОГО ЭЛЕМЕНТА (PIVOT):

1. Выбор первого элемента (quicksort_first):
   - Если входной массив уже отсортирован (1, 2, 3...) или отсортирован в обратном порядке (...3, 2, 1),
     алгоритм работает за O(n^2). 
   - Это происходит потому, что на каждом шаге мы выбираем самый минимальный (или максимальный) 
     элемент. В итоге дерево рекурсии вырождается в список: вместо разбиения массива пополам 
     мы просто «откусываем» по одному элементу за раз.

2. Выбор середины (quicksort_middle):
   - Решает проблему уже отсортированных массивов. В этом случае разбиение будет 
     сбалансированным (примерно пополам), и сложность составит O(n log n).

3. Случайный выбор (quicksort_random):
   - Лучшая стратегия для защиты от «плохих» входных данных. Почти невозможно 
     подобрать такой массив специально, чтобы алгоритм всегда выбирал худший pivot.
"""

def quicksort_first(arr):
    if len(arr) <= 1: return arr
    pivot = arr[0]
    left = [x for x in arr[1:] if x <= pivot]
    right = [x for x in arr[1:] if x > pivot]
    return quicksort_first(left) + [pivot] + quicksort_first(right)

def quicksort_middle(arr):
    if len(arr) <= 1: return arr
    mid = len(arr) // 2
    pivot = arr[mid]
    others = arr[:mid] + arr[mid+1:]
    left = [x for x in others if x <= pivot]
    right = [x for x in others if x > pivot]
    return quicksort_middle(left) + [pivot] + quicksort_middle(right)

def quicksort_random(arr):
    if len(arr) <= 1: return arr
    rand_idx = random.randint(0, len(arr) - 1)
    pivot = arr[rand_idx]
    others = arr[:rand_idx] + arr[rand_idx+1:]
    left = [x for x in others if x <= pivot]
    right = [x for x in others if x > pivot]
    return quicksort_random(left) + [pivot] + quicksort_random(right)

# Тесты
data = [3, 6, 8, 10, 1, 2, 1]
print("First pivot:", quicksort_first(data))
print("Middle pivot:", quicksort_middle(data))
print("Random pivot:", quicksort_random(data))

# Демонстрация O(n^2) для quicksort_first
sorted_data = list(range(1000))  # На очень больших массивах quicksort_first упадет с RecursionError
````

````python
#3. 
import time
from functools import wraps

def timer(func):
    """Декоратор, измеряющий время выполнения функции."""
    @wraps(func) #нужно чтобы функция после  декорирования не забыла свое имя
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()  # Начало отсчета
        result = func(*args, **kwargs)    # Выполнение функции
        end_time = time.perf_counter()    # Конец отсчета
        
        execution_time = end_time - start_time
        print(f"⏱ Функция '{func.__name__}' выполнена за {execution_time:.6f} сек.")
        return result
    return wrapper

# Пример использования:
@timer
def some_heavy_task(n):
    """Функция, имитирующая долгую работу."""
    total = 0
    for i in range(n):
        total += i**2
    return total

# Запуск
some_heavy_task(10_000_000)
````

<details>
<summary><b>Ответ 1: Global Interpreter Lock (GIL)</b></summary>

**GIL** — это мьютекс, который разрешает только одному потоку выполнять байткод Python в один момент времени.
Из-за этого модуль `threading` не ускорит CPU-bound задачи (математику, расчеты). Треды будут просто бороться за GIL.
**Как обойти:** 
1. Использовать `multiprocessing` (создаёт изолированные процессы, у каждого свой GIL).
2. Использовать C-расширения или библиотеки типа NumPy, которые выносят вычисления из Python и отпускают GIL.
</details>

<details>
<summary><b>Ответ 2: Garbage Collector и Reference Counting</b></summary>

Python управляет памятью в два этапа:
1. **Reference Counting (Счетчик ссылок):** На каждый объект в памяти ведется счетчик переменных, которые на него ссылаются. Если счетчик падает до 0 — объект сразу удаляется.
2. **Garbage Collector (Сборщик мусора):** Нужен для поиска **циклических ссылок** (когда объект A ссылается на B, а B — на A). В этом случае счетчик ссылок никогда не станет нулем. GC (модуль `gc`) периодически сканирует графы объектов и удаляет "острова", к которым нет доступа из основной программы.
</details>

**Решение:** 
- Итерироваться по копии словаря: `for k, v in list(d.items()):`
- Итерироваться по ключам: `for k in list(d.keys()):`
</details>

<details>
<summary><b>Ответ 4: Deep Copy vs Shallow Copy</b></summary>

Допустим, `list_a = [[1, 2], [3, 4]]`.
- **Shallow Copy (`copy.copy()`):** Создаст новый внешний список, но внутри него будут лежать ссылки на те же самые внутренние списки `[1, 2]` и `[3, 4]`. При изменении внутреннего списка в оригинале, копия тоже изменится.
- **Deep Copy (`copy.deepcopy()`):** Рекурсивно создает копию всех вложенных объектов. Они становятся полностью независимыми структурами.
</details>

<details>
<summary><b>Ответ 5: Two Sum и Group Anagrams</b></summary>

**Two Sum:** Проходим по числам массива. В словаре сохраняем разницу `target - num` и индекс числа `num`. Если текущее `num` есть в словаре, мы нашли пару.
**Group Anagrams:** Проходим по строкам. Сортируем каждую строку (`''.join(sorted(s))`) и используем отсортированную строку как ключ словаря (или `collections.defaultdict`), а саму строку добавляем в список по этому ключу.
</details>

## 🧠 Частые дополнительные вопросы (на понимание)

<details>
<summary><b>1. Разница между `__new__` и `__init__`</b></summary>

- `__new__` — это статический метод, который **создает и возвращает** новый экземпляр класса (аллоцирует память). Вызывается самым первым.
- `__init__` — это метод, который **инициализирует** (заполняет атрибутами) уже созданный экземпляр, переданный параметром `self`. Ничего не возвращает.
Чаще всего `__new__` переопределяют для реализации паттерна Singleton.
</details>

<details>
<summary><b>2. Multithreading vs Multiprocessing — когда что использовать?</b></summary>

- **Multithreading (Потоки):** Использовать для **I/O-Bound** задач (сетевые запросы, скачивание файлов, запись в БД). GIL не мешает, когда поток ждет ответа по сети.
- **Multiprocessing (Процессы):** Использовать для **CPU-Bound** задач (машинное обучение на CPU, рендер изображений, сортировка огромных массивов). 
- **AsyncIO:** Ещё лучше использовать для масштабных I/O задач, так как это асинхронное выполнение в одном потоке, потребляющее минимум ресурсов.
</details>

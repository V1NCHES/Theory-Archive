## Блок 1: Базовые типы и Изменяемость (Mutability)

### 🔎 Теория
В Python все является объектом. Важное деление объектов — на **изменяемые (mutable)** и **неизменяемые (immutable)**.

1. **Immutable (Неизменяемые):** `int`, `float`, `str`, `tuple`, `frozenset`, `bytes`.
   - При попытке изменить объект создается его копия с новым `id`.
2. **Mutable (Изменяемые):** `list`, `dict`, `set`, `bytearray`.
   - Можно менять "на месте" (in-place).

```python
# Пример изменяемости
a = [1, 2, 3]
b = a
b.append(4)
print(f"Список a: {a} (id: {id(a)})") # a изменился тоже!
```

## Блок 2: Срезы, Comprehensions и Генераторы

- **Срезы:** `lst[start:stop:step]`.
- **Генераторы:** Не хранят все данные в памяти, а выдают их по запросу (`next()` или `for`).

```python
# List comprehension vs Generator
n = 1000
list_comp = [x**2 for x in range(n)]
gen_exp = (x**2 for x in range(n))
print(type(gen_exp))
```

## Блок 3: Функции: *args, **kwargs и Замыкания

- `*args` — пакует позиционные аргументы в кортеж.
- `**kwargs` — пакует именованные аргументы в словарь.
- **Замыкание:** Внутренняя функция хранит состояние из внешней области видимости.

```python
def power(exponent):
    def inner(base):
        return base ** exponent
    return inner

square = power(2)
print(square(5)) # 25
```

## Блок 4: Исключения

Иерархия исключений: `BaseException` -> `Exception` -> ...
- Свои классы ошибок должны наследоваться от `Exception`.

```python
class MyCustomError(Exception): pass

try:
    raise MyCustomError("Что-то пошло не так!")
except MyCustomError as e:
    print(f"Поймали: {e}")
```

## Блок 5: ООП (Инкапсуляция, MRO, super)

- **MRO (Method Resolution Order):** Порядок поиска методов при множественном наследовании.
- **super():** Вызывает следующий класс в цепочке MRO.

```python
class Base:
    def say(self): print("Base")

class childA(Base):
    def say(self): print("A"); super().say()

class childB(Base):
    def say(self): print("B"); super().say()

class GrandChild(childA, childB):
    def say(self): print("Grand"); super().say()

GrandChild().say() # Обратите внимание на порядок!
```

## Блок 6: Спец-методы (Dunder methods)

`__str__` vs `__repr__`:
- `__str__` — для пользователя (читаем).
- `__repr__` — для разработчика (отладка, должен быть однозначным).

```python
class Person:
    def __init__(self, name): self.name = name
    def __repr__(self): return f"Person('{self.name}')"
    def __str__(self): return f"Человек по имени {self.name}"

p = Person("Иван")
print(str(p))
print(repr(p))
```

## Блок 7: Алгоритмы (Бинарный поиск и Сортировка)

- **Бинарный поиск:** O(log n). Массив должен быть отсортирован.
- **Сортировка пузырьком:** O(n^2). Простейшая, но неэффективная.

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target: return mid
        elif arr[mid] < target: low = mid + 1
        else: high = mid - 1
    return -1

data = [1, 3, 5, 7, 9, 11]
print(f"Индекс 7: {binary_search(data, 7)}")
```

## Блок 8: Структуры данных (Стек, Очередь, Хеш-таблица)

- **Стек:** LIFO (Last In, First Out). `append()` / `pop()`.
- **Очередь:** FIFO (First In, First Out). `collections.deque`.
- **Хеш-таблица:** Лежит в основе `dict` и `set`. Обеспечивает O(1) в среднем для поиска.

```python
from collections import deque

queue = deque(["A", "B", "C"])
queue.append("D")
print(queue.popleft()) # "A"
print(queue) # ["B", "C", "D"]
```

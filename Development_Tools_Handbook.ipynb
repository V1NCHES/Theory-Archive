## Блок 1: Работа с Git и Методология GitFlow

### 🔎 Основные команды
- `git init` — создать репозиторий.
- `git clone <url>` — скопировать удаленный репозиторий.
- `git add .` — подготовить изменения к коммиту.
- `git commit -m "сообщение"` — сохранить изменения.
- `git push` — отправить изменения в облако.
- `git pull` — забрать изменения из облака и объединить.
- `git fetch` — забрать изменения из облака (без объединения).

### 🌊 Методология GitFlow
GitFlow — это схема работы с ветками, которая помогает не ломать основной код.
1. **main (или master):** Стабильный код, который уже работает (продуктив).
2. **develop:** Основная ветка разработки. Все новые функции попадают сюда.
3. **feature/имя-задачи:** Ветка для разработки конкретной фичи. Создается от `develop`, вливается в `develop` через Pull Request (PR).
4. **hotfix/имя-ошибки:** Срочное исправление багов в `main`.

## Блок 2: Виртуальные окружения (venv) и pip

### 🔎 Теория
**Зачем это нужно?** У разных проектов могут быть разные версии библиотек (например, Pandas 1.0 и Pandas 2.0). Если устанавливать их глобально, возникнет конфликт. `venv` создает изолированную папку с Python.

- Создание: `python -m venv venv`.
- Активация (Win): `venv\Scripts\activate`.
- Заморозка зависимостей: `pip freeze > requirements.txt`.
- Установка из файла: `pip install -r requirements.txt`.

```python
import sys
print(f"Путь к текущему интерпретатору: {sys.executable}")
# Если путь содержит 'venv', значит вы работаете в виртуальном окружении!
```

## Блок 3: Продвинутый Jupyter (Магические команды)

Магические команды в Jupyter начинаются с `%` (для одной строки) или `%%` (для всей ячейки).

- `%timeit` — замерить время выполнения строки.
- `%%time` — замерить время выполнения всей ячейки.
- `%ls` — список файлов в папке.
- `%matplotlib inline` — отображение графиков прямо в ноутбуке.

```python
import time

%%time
# Замеряем время работы цикла
sum([x**2 for x in range(1000000)])
```

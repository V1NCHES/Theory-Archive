## Блок 1: Работа с Git и Методология GitFlow

### 🔎 Основные команды
- `git init` — создать репозиторий.
- `git clone <url>` — скопировать удаленный репозиторий.
- `git add .` — подготовить изменения к коммиту.
- `git commit -m "сообщение"` — сохранить изменения.
- `git push` — отправить изменения в облако.
- `git pull` — забрать изменения из облака и объединить.
- `git fetch` — забрать изменения из облака (без объединения).

### 🌊 Методология GitFlow
GitFlow — это схема работы с ветками, которая помогает не ломать основной код.
1. **main (или master):** Стабильный код, который уже работает (продуктив).
2. **develop:** Основная ветка разработки. Все новые функции попадают сюда.
3. **feature/имя-задачи:** Ветка для разработки конкретной фичи. Создается от `develop`, вливается в `develop` через Pull Request (PR).
4. **hotfix/имя-ошибки:** Срочное исправление багов в `main`.

## Блок 2: Виртуальные окружения (venv) и pip

### 🔎 Теория
**Зачем это нужно?** У разных проектов могут быть разные версии библиотек (например, Pandas 1.0 и Pandas 2.0). Если устанавливать их глобально, возникнет конфликт. `venv` создает изолированную папку с Python.

- Создание: `python -m venv venv`.
- Активация (Win): `venv\Scripts\activate`.
- Заморозка зависимостей: `pip freeze > requirements.txt`.
- Установка из файла: `pip install -r requirements.txt`.

```python
import sys
print(f"Путь к текущему интерпретатору: {sys.executable}")
# Если путь содержит 'venv', значит вы работаете в виртуальном окружении!
```

## Блок 3: Продвинутый Jupyter (Магические команды)

Магические команды в Jupyter начинаются с `%` (для одной строки) или `%%` (для всей ячейки).

- `%timeit` — замеряет время выполнения строки.
- `%%time` — замеряет время выполнения всей ячейки.
- `%ls` — список файлов в папке.
- `%matplotlib inline` — отображение графиков прямо в ноутбуке.

```python
%%time
# Замеряем время работы цикла
sum([x**2 for x in range(1000000)])
```

### 📝 Задача 2
Используй `%timeit`, чтобы сравнить скорость вычисления суммы квадратов чисел от 1 до 1000 через цикл `for` и через функцию `sum()` с генератором.

```python
# Первый тест
print("Цикл for:")
%timeit total = 0; total = sum(range(1, 1000))

# Второй тест  
print("\nГенератор списка:")
%timeit sum([x for x in range(1,1000)])
```

### ✅ Ответы на вопросы

<details>
<summary><b>Ответ 1: Merge vs Rebase и политики работы с ветками</b></summary>

**В чем разница:**
- `Merge` создает новый комит слияния, сохраняя всю историю изменений в том виде, в котором они были произведены. Это делает историю «честной», но перегруженной комитами слияния.
- `Rebase` «переписывает» историю, перенося ваши комиты поверх последних изменений целевой ветки. Это создает чистую линейную историю, но изменяет хеши комитов.

**Почему запрещают `push --force` в main:**
- `push --force` перезаписывает историю на сервере. Если кто-то другой уже скачал старую историю, его локальный репозиторий «сломается», и возникнет хаос при попытке синхронизации. В `main` история должна быть неизменной.

**Squash + Merge vs Rebase:**
- В крупных компаниях предпочитают `Squash + Merge`, так как он объединяет всю работу из ветки в один чистый комит в `main`. Это проще контролировать, чем длинный `rebase`, который может потребовать пошагового разрешения конфликтов для каждого комита.
</details>

<details>
<summary><b>Ответ 2: Разрешение конфликтов (Merge Conflicts)</b></summary>

**Конфликт слияния** возникает, когда два человека изменили одни и те же строки кода в одном файле, и Git не может понять, чей вариант правильнее.

**Пошаговый план разрешения в PR:**
1. Перейти в свою ветку локально: `git checkout feature/your-branch`.
2. Стянуть последние изменения из целевой ветки: `git pull origin develop`.
3. Git сообщит о конфликтах. Открыть конфликтные файлы в IDE.
4. Вручную выбрать нужные строки между заголовками `<<<<<<<` и `>>>>>>>`.
5. Сохранить файлы и добавить их в индекс: `git add <filename>`.
6. Завершить слияние: `git commit`.
7. Отправить исправленную версию в облако: `git push`.
</details>

<details>
<summary><b>Ответ 3: Инцидент безопасности (пароли в Git)</b></summary>

**Критический план действий:**
1. **Смена учетных данных:** Прямо сейчас смените все пароли/ключи, которые были в файле. Удаление из Git НЕ делает секреты безопасными, так как они остаются в истории.
2. **Очистка истории:** Используйте специальные утилиты (например, `BFG Repo-Cleaner` или `git filter-repo`), чтобы полностью вычистить файл `.env` из всех комитов во всей истории.
3. **Обновление удаленного репозитория:** Сделайте `git push --force`, чтобы перезаписать историю на GitHub/GitLab.
4. **Профилактика:** Добавьте `.env` в `.gitignore` и настройте использование переменных окружения (Environment Variables) через CI/CD или Secret Manager (Vault).
</details>

<details>
<summary><b>Ответ 4: Фиксация версий в requirements.txt</b></summary>

**Зачем это нужно:**
1. **Воспроизводимость (Deterministic Builds):** Чтобы код, который работает у вас, точно так же работал у коллеги и на сервере. Без фиксации `pip` может поставить новую версию библиотеки, в которой изменился синтаксис (Breaking Changes), и проект упадет.
2. **Избежание сюрпризов:** Если библиотека `pandas` обновится с 2.0 до 3.0 за ночь, ваш код может сломаться без вашего участия. Фиксация версии гарантирует стабильность.
3. **Безопасность:** Вы уверены, что используете именно ту версию, которую проверили на уязвимости.
</details>

## 🧠 Частые дополнительные вопросы (на понимание)

<details>
<summary><b>1. Docker Image vs Docker Container</b></summary>

- **Image (Образ):** Это неизменяемый шаблон или "чертеж". В нем лежит кодовая база, зависимости и ОС. Образ нельзя запустить, из него можно создать контейнер.
- **Container (Контейнер):** Это живой экземпляр образа (процесс ОС). У него есть оперативная память, доступ к диску и время жизни. Как класс (Image) и объект (Container) в ООП.
</details>

<details>
<summary><b>2. Базовые команды и права Linux</b></summary>

Для чего нужна команда `chmod 777 file.txt`?
- **Ужасная с точки зрения безопасности** команда. Расшифровывается как rwxrwxrwx (Read, Write, Execute для Владельца, Группы и Остальных).
- Чтение = 4, Запись = 2, Выполнение = 1.
- 7 (4+2+1) означает полный доступ для всех. 
Обычно разработчики используют `chmod 755` или `644` в целях безопасности.
</details>

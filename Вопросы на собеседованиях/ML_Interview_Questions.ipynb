# 🎓 Подготовка к собеседованию: TOP-25 вопросов по Data Science & ML

В этом ноутбуке собраны развернутые ответы на популярные вопросы с собеседований, структурированные для глубокого понимания.

---

## Часть 1: Основы ML и Классические алгоритмы

### 1. Как работает логистическая регрессия?
**Ответ:** Логистическая регрессия — это алгоритм классификации (несмотря на название). 
- **Суть:** Мы строим линейную комбинацию признаков $z = w^T x + b$. Чтобы превратить это число в вероятность, мы пропускаем $z$ через **сигмоиду**: $\sigma(z) = \frac{1}{1 + e^{-z}}$. Результат всегда лежит в диапазоне [0, 1].
- **Обучение:** Минимизируется функция потерь **Log-Loss** (бинарная кросс-энтропия) с помощью градиентного спуска. Она сильно штрафует модель за уверенные неправильные предсказания.
- **Интерпретация:** Веса $w$ показывают, как изменение признака влияет на логарифм шансов (log-odds).

### 2. Что такое переобучение (Overfitting) и как его заметить?
**Ответ:** Переобучение — это ситуация, когда модель "выучила" тренировочные данные вместе с шумом и потеряла обобщающую способность.
- **Как заметить:** Огромный разрыв в метриках: на трейне точность 99%, а на тесте — 70%.
- **Методы борьбы (фишки):**
    1. **Data Augmentation:** Искусственное увеличение выборки (повороты фото, синонимы в тексте).
    2. **SMOTE:** Синтетическая генерация примеров редкого класса (для борьбы с дисбалансом).
    3. **Добавление шума:** Помогает модели быть устойчивее и не зацикливаться на точных значениях признаков.
    4. **Регуляризация:** Штраф за большие веса.

### 3. Почему L1-регуляризация зануляет признаки?
**Ответ:** L1 (Lasso) добавляет к лоссу слагаемое $\lambda \sum |w_i|$. 
- **Геометрическое объяснение:** Линии уровня функции потерь L1 — это ромбы (в 2D). Минимум функции чаще всего находится в вершинах ромба, которые лежат на осях координат. На осях одна из координат равна нулю — так признак исключается из модели.
- **Зачем это нужно:** Автоматический отбор признаков (Feature Selection) и борьба с мультиколлинеарностью.

### 4. Как устроено дерево решений?
**Ответ:** Дерево — это иерархическая структура из правил "если-то".
- **Разбиение:** В каждом узле мы выбираем признак и порог, которые дают максимальный **прирост информации (Information Gain)** или максимальное снижение **критерия Джини / Энтропии**.
- **Жадный алгоритм:** На каждом шаге дерево ищет лучшее разбиение здесь и сейчас, не заглядывая в будущее.

### 5. Переобучение в деревьях и прунинг (Pruning)?
**Ответ:** Одиночное дерево очень склонно к переобучению (может построить лист под каждый объект).
- **Прунинг (стрижка):** Это процесс удаления узлов, которые не приносят значимого улучшения предсказаний. Бывает **pre-pruning** (остановка роста по макс. глубине или мин. числу объектов в листе) и **post-pruning** (удаление веток уже после построения дерева).

### 6. Как работает Случайный Лес (Random Forest)?
**Ответ:** Это ансамбль независимых деревьев, использующий метод **Bagging**.
- **Принципы:** 
    1. **Bootstrapping:** Каждое дерево обучается на случайной подвыборке с возвращением.
    2. **Feature Randomness:** При каждом разбиении выбирается случайное подмножество признаков (обычно $\sqrt{d}$).
- **Результат:** Деревья получаются разными (декоррелированными), а их усреднение резко снижает **дисперсию (Variance)** всей модели.

### 7. Чем отличается Бустинг от Random Forest?
**Ответ:** 
- **RF:** Деревья строятся параллельно и независимо. Снижает Variance.
- **Бустинг:** Деревья строятся последовательно. Каждое новое дерево обучается исправлять ошибки (антиградиент функции потерь) предыдущих. Снижает **смещение (Bias)**.
- **Нюанс:** Бустинг обычно точнее, но капризнее к гиперпараметрам и легче переобучается.

### 8. Метрики регрессии и классификации?
**Ответ:** 
- **Регрессия:** 
    - **MSE:** Квадратичная ошибка (чувствительна к выбросам).
    - **MAE:** Абсолютная ошибка (более устойчива).
    - **R^2:** Доля объясненной дисперсии.
- **Классификация:** 
    - **Accuracy:** Общая точность (бесполезна при дисбалансе).
    - **Log-Loss:** Оценивает уверенность модели (вероятности).
    - **F1-Score:** Гармоническое среднее Precision и Recall.

### 9. Что такое ROC-AUC и как его интерпретировать?
**Ответ:** Это площадь под кривой зависимости True Positive Rate от False Positive Rate.
- **Интерпретация:** Вероятность того, что модель поставит случайному объекту класса 1 скор выше, чем случайному объекту класса 0. 
- **Значения:** 1.0 — идеальная модель, 0.5 — случайное угадывание.

### 10. Precision — как объяснить на собеседовании?
**Ответ:** "Представьте, что модель — это полиция. Precision — это ответ на вопрос: 'Из всех, кого мы арестовали, сколько реально были преступниками?'. Если Precision низкий, мы арестовываем много невиновных (много False Positives)".

### 11. Recall и его применение в медицине?
**Ответ:** Recall (полнота) — это ответ на вопрос: 'Сколько из всех реальных преступников нам удалось поймать?'. В медицине это критично: нам нельзя пропустить больного (False Negative), даже если мы ошибочно отправим здорового на доп. обследование. Лучше перебдеть.

### 12. Связь Recall и ROC-AUC?
**Ответ:** ROC-кривая строится по точкам (FPR, TPR). Помним, что **TPR = Recall**. То есть ROC-AUC показывает, как меняется Recall (полнота) при изменении порога вероятности и сопутствующем росте ложных тревог (FPR).

## Часть 2: Продвинутые инструменты и Вероятность

### 13. Зачем нужна PR-кривая и в чём её отличие от ROC-AUC?
**Ответ:** PR-кривая (Precision-Recall) используется при **сильном дисбалансе классов** (когда объектов класса 1 очень мало). 
- ROC-AUC может быть обманчиво высоким, если объектов 0-класса миллионы.
- PR-кривая фокусируется только на миноритарном классе и нагляднее показывает качество модели в таких условиях.

### 14. Подбор гиперпараметров: GridSearch, RandomSearch, Optuna?
**Ответ:** 
- **GridSearch:** Тупой перебор всех комбинаций. Долго, но надежно (в рамках сетки).
- **RandomSearch:** Случайный выбор комбинаций. Обычно находит хорошие параметры быстрее, чем Grid.
- **Optuna:** Байесовская оптимизация. Она "думает": запоминает результаты прошлых попыток и пробует те параметры, которые с большей вероятностью улучшат метрику. Самый современный стандарт.

### 15. Как оценить важность признаков (Feature Importance)?
**Ответ:** 
1. **Встроенные (в деревья):** Сколько раз признак использовался для разбиения или как сильно он снизил Джини.
2. **Permutation Importance:** Мы случайно перемешиваем значения одного признака в тесте. Если метрика модели упала — признак был важным.
3. **SHAP / LIME:** Математически обоснованные методы (из теории игр), которые показывают вклад каждого признака в конкретное предсказание.

### 16. Работа с категориальными фичами и CatBoost?
**Ответ:** Стандарт — One-Hot или Label encoding. 
- **CatBoost:** Его фишка в том, что он умеет работать с категориями "из коробки". Он использует **Ordered Target Encoding** (вычисляет статистику по таргету, учитывая порядок данных), что предотвращает утечку данных (data leakage).

### 17. В чём разница между `==` и `is` в Python?
**Ответ:** 
- `==` проверяет **значение** (эквивалентность). `[1,2] == [1,2]` даст True.
- `is` проверяет **идентичность** (ссылаются ли переменные на один и тот же объект в памяти). `[1,2] is [1,2]` даст False.

### 18. Как устроены словари (`dict`) и что такое коллизии?
**Ответ:** Словарь — это хэш-таблица.
- Ключ хэшируется в число, которое определяет индекс в массиве.
- **Коллизия:** Когда разные ключи дают один хэш-индекс. В Python решается методом открытой адресации (ищется следующая свободная ячейка).

### 19. Что такое итераторы и зачем они нужны?
**Ответ:** Итератор — это объект, который выдает элементы по одному (лениво). 
- **Зачем:** Экономия памяти. Если у нас файл на 10 ГБ, мы не читаем его целиком (`list`), а обрабатываем построчно с помощью итератора.

### 20. Теорвер: вероятность суммы 11 при броске двух кубиков?
**Ответ:** Всего комбинаций $6 \times 6 = 36$. Сумму 11 дают пары (5,6) и (6,5) — их 2 штуки. 
- Вероятность: $P = 2/36 = 1/18 \approx 5.5\%$.

### 21. Теорема Байеса: как выглядит и как применять?
**Ответ:** $P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$. 
- Позволяет обновить вероятность гипотезы $A$ при получении новых данных $B$. Используется в спам-фильтрах, медицине и A/B тестах.

### 22. Что такое p-value и как его объяснить?
**Ответ:** Это вероятность получить такие или еще более экстремальные результаты, **если нулевая гипотеза верна**. 
- Если p-value < 0.05, мы говорим: "Слишком маловероятно, что это случайность, значит, эффект реален".

### 23. Какие стат-тесты бывают и когда их использовать?
**Ответ:** 
- **T-test:** Сравнение средних в двух группах (если данные распределены нормально).
- **Mann-Whitney:** Сравнение распределений (если данные не нормальные, работаем с рангами).
- **Chi-Square (Хи-квадрат):** Для категориальных данных (есть ли зависимость между полом и покупкой).

### 24. Функции активации в нейросетях: зачем и какие?
**Ответ:** Вносят нелинейность. 
- **Sigmoid/Tanh:** Сейчас редки в скрытых слоях (проблема затухающих градиентов).
- **ReLU:** Стандарт. Пропускает положительные значения как есть, отрицательные зануляет. Быстро считается.
- **Softmax:** На последнем слое классификации (сумма выходов = 1).

### 25. Что делает Dropout при инференсе (тестировании)?
**Ответ:** **Ничего.** 
- Dropout работает только при обучении, чтобы нейроны не привыкали к присутствию друг друга.
- При инференсе все нейроны работают, но их веса обычно масштабируются (умножаются на вероятность сохранения $p$), чтобы сохранить средний масштаб сигнала.

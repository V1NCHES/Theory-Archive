# 🎓 Вопросы на собеседованиях по Machine Learning

Этот ноутбук содержит вопросы и ответы с YouTube-канала (согласно видео из плейлиста).

---

## Часть 1: Базовые концепции (Видео 1)

### 1. Что такое Precision и Recall? Какая метрика важнее?
**Ответ:** Precision (точность) — это доля правильно предсказанных положительных объектов среди всех объектов, которые модель назвала положительными. Recall (полнота) — это доля правильно предсказанных положительных объектов среди всех реально положительных объектов. Важность зависит от задачи: в медицине (диагностика болезней) важнее Recall, в антиспаме — Precision.

### 2. Что такое Overfitting и Underfitting?
**Ответ:** Overfitting (переобучение) — это когда модель слишком хорошо выучила тренировочные данные (шум), но плохо работает на новых данных. Underfitting (недообучение) — когда модель слишком простая и не может уловить зависимости даже в тренировочных данных.

### 3. В чем разница между L1 и L2 регуляризацией?
**Ответ:** L1 (Lasso) добавляет к функции потерь сумму абсолютных значений весов, что приводит к занулению некоторых весов (отбор признаков). L2 (Ridge) добавляет сумму квадратов весов, что уменьшает их значения, но не обнуляет их полностью.

### 4. Что такое Bias-Variance Tradeoff?
**Ответ:** Это баланс между ошибкой смещения (Bias) и ошибкой разброса (Variance). Простые модели имеют высокий Bias и низкий Variance. Сложные модели — низкий Bias и высокий Variance. Цель — найти золотую середину.

### 5. Зачем нужна кросс-валидация?
**Ответ:** Для более надежной оценки качества модели. Данные делятся на K частей (фолдов), и модель обучается K раз, каждый раз используя новый фолд для валидации, а остальные — для обучения.

### 6. Как работает градиентный спуск?
**Ответ:** Это итеративный алгоритм оптимизации. Мы считаем производную функции потерь по весам и делаем шаг в сторону, противоположную градиенту, чтобы минимизировать ошибку.

### 7. Random Forest vs Gradient Boosting: основные отличия?
**Ответ:** Random Forest строит деревья независимо и усредняет результат (бэггинг). Boosting строит деревья последовательно, где каждое следующее дерево исправляет ошибки предыдущих.

### 8. Как обрабатывать пропущенные значения?
**Ответ:** Можно удалить строки/столбцы (если их мало), заменить на среднее/медиану/моду, использовать константу или обучить модель для предсказания пропусков.

### 9. Зачем делать Feature Scaling?
**Ответ:** Чтобы признаки с большими значениями не доминировали над другими. Это критично для метрических алгоритмов (KNN, SVM) и градиентного спуска.

### 10. One-Hot Encoding vs Label Encoding?
**Ответ:** Label Encoding присваивает категориям числа (0, 1, 2...). One-Hot создает новые столбцы-флаги для каждой категории. One-Hot лучше для номинальных данных, где нет порядка.

### 11. Метрики регрессии vs Классификации?
**Ответ:** Регрессия: MSE, MAE, R^2, RMSE. Классификация: Accuracy, Precision, Recall, F1-score, Logloss, ROC-AUC.

### 12. Что такое Проклятие размерности (Curse of Dimensionality)?
**Ответ:** С ростом количества признаков объем пространства растет экспоненциально, и данные становятся разреженными. Это усложняет поиск закономерностей и требует больше данных.

## Часть 2: Алгоритмы и Нейросети (Видео 2)

### 13. Как работает логистическая регрессия?
**Ответ:** Она предсказывает вероятность принадлежности к классу. Линейная комбинация признаков пропускается через сигмоиду (Sigmoid), которая переводит любое число в диапазон [0, 1].

### 14. Что такое Kernel Trick в SVM?
**Ответ:** Это способ перевести данные в пространство более высокой размерности, где они становятся линейно разделимыми, без явного вычисления координат в этом пространстве.

### 15. Как работает KNN? Как выбрать K?
**Ответ:** Алгоритм ищет K ближайших соседей к объекту и присваивает ему класс большинства. K обычно выбирается через кросс-валидацию (нечетное число).

### 16. Зачем нужен PCA (метод главных компонент)?
**Ответ:** Для снижения размерности данных с сохранением максимальной дисперсии. Помогает визуализировать данные и убрать шум.

### 17. Как работает K-means? Как выбрать число кластеров?
**Ответ:** Итеративно обновляет центроиды кластеров. Число кластеров выбирается методом "локтя" (Elbow method) или через силуэт (Silhouette score).

### 18. Что показывает ROC-AUC?
**Ответ:** Это площадь под кривой ошибок (TPR vs FPR). Она показывает способность модели разделять классы независимо от выбранного порога вероятности.

### 19. Что такое ансамбли (Ensembles)?
**Ответ:** Объединение нескольких моделей для получения более качественного результата. Основные типы: бэггинг, бустинг и стекинг.

### 20. Как бороться с несбалансированными классами?
**Ответ:** Ресемплирование (Oversampling/Undersampling), использование правильных метрик (не Accuracy), изменение весов классов в функции потерь.

### 21. Что такое Backpropagation?
**Ответ:** Алгоритм обратного распространения ошибки. Мы считаем градиент функции потерь от конца сети к началу, используя цепное правило (chain rule), чтобы обновить веса.

### 22. Зачем нужны функции активации (ReLU, Sigmoid, Tanh)?
**Ответ:** Чтобы внести нелинейность в нейронную сеть. Без них сеть любой глубины была бы просто большой линейной регрессией.

### 23. Что такое Dropout?
**Ответ:** Метод регуляризации нейросетей, при котором на каждой итерации обучения случайно "выключается" часть нейронов, что предотвращает переобучение.

### 24. Что такое Batch Normalization?
**Ответ:** Слой, который нормализует выходы предыдущего слоя (среднее 0, дисперсия 1) внутри каждого батча. Ускоряет обучение и делает его более стабильным.

### 25. Разница между CNN и RNN?
**Ответ:** CNN (сверточные сети) хороши для пространственных данных (изображения). RNN (рекуррентные сети) предназначены для последовательностей (текст, временные ряды), так как имеют "память" о предыдущих шагах.
